```python
import pandas as pd

df_csv = pd.read_csv(r"D:\Mechaine_Learning_Journey\Quality of life data.csv")
print("Columns in your DataFrame:")
print(df_csv.columns.tolist())
print("\nMissing values in each column:")
print(df_csv.isnull().sum())
if 'age_at_death' in df_csv.columns:
    df_csv['age_at_death'] = df_csv['age_at_death'].fillna(df_csv['age_at_death'].mean())
    print("\nFilled missing 'age_at_death' with mean (if any).")
numeric_cols = df_csv.select_dtypes(include=['float64', 'int64']).columns
df_csv[numeric_cols] = df_csv[numeric_cols].fillna(df_csv[numeric_cols].mean())
df_csv.dropna(how='all', inplace=True)
print("\nAfter cleaning:")
print(df_csv.head())
print("\nShape:", df_csv.shape)
print("\nMissing values now:\n", df_csv.isnull().sum())
```
